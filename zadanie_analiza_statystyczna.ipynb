```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Wczytaj dane
dataset = pd.read_csv("C:/Users/abero/Downloads/titanic.csv")
```

```python
# Podstawowe informacje
print(dataset['Fare'].describe())
```

```python
mean_fare = dataset['Fare'].mean()
median_fare = dataset['Fare'].median()
q1 = dataset['Fare'].quantile(0.25)
q3 = dataset['Fare'].quantile(0.75)

print(f"Średnia opłata (Fare): {mean_fare:.2f}")
print(f"Mediana opłaty: {median_fare:.2f}")
print(f"1. kwartyl (Q1): {q1:.2f}")
print(f"3. kwartyl (Q3): {q3:.2f}")
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x=dataset['Fare'])
plt.title('Wykres pudełkowy zmiennej Fare')
plt.xlabel('Fare')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(dataset['Fare'], bins=30, color='skyblue', edgecolor='black')
plt.axvline(mean_fare, color='red', linestyle='dashed', linewidth=1.5, label='Średnia')
plt.axvline(median_fare, color='green', linestyle='dashed', linewidth=1.5, label='Mediana')
plt.title('Histogram zmiennej Fare')
plt.xlabel('Fare')
plt.ylabel('Liczba pasażerów')
plt.legend()
plt.show()
```
